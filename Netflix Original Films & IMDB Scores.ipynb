```python
import pandas as pd
import matplotlib.pyplot as plt
```

### About this dataset
Netflix Original Films & IMDB Scores as of 06/01/21

```python
df = pd.read_csv("./NetflixOriginals.csv")

df
```

### Best IMDB Score

```python
best_score = df.sort_values(by = "IMDB Score" , ascending=False)

#First 10 IMBD Score
best_score.head(10)
```

### Movie Genre

```python
genre_unique_counts = df.groupby("Genre")["Genre"].value_counts()
genre_unique_name = df["Genre"].unique()


plt.figure(figsize=(18,9))
plt.bar(genre_unique_name,genre_unique_counts)
plt.ylabel("Genre Counts")
plt.xticks(rotation="vertical")
plt.xlabel("Genre Name")
plt.show()
```

```python
#Genre Counts
df["Genre"].value_counts()
```

### Highest Runtime

```python
high_runtime = df.sort_values(by = "Runtime" ,ascending=False )

#Top 10 Runtime
high_runtime.head(10)
```

### Most Popular Movie Language

```python
language = df["Language"].unique()
count_language = df.groupby("Language")["Language"].value_counts()

plt.figure(figsize=(20,8))
plt.bar(language,count_language)
plt.ylabel("Number")
plt.xticks(rotation="vertical")
plt.xlabel("Languages")
plt.show()
```
