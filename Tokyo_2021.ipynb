## Medals Count of Countries

### Cleaning Data

```python
columns = df.columns
cleaning_columns = []
for i in columns:
    k = i.replace(" ","_")
    df.rename(columns = {i:k},inplace = True)
```

### Gold Medals


```python
gold = df[df["Gold_Medal"]>0]


plt.figure(figsize=(15,8))
plt.bar(gold["Country"],gold["Gold_Medal"])
plt.xticks(rotation = "vertical")
plt.xlabel("Country")
plt.ylabel("Value")
plt.show()
```

### Silver Medals


```python
silver = df[df["Silver_Medal"]>0]

plt.figure(figsize=(15,8))
plt.bar(silver["Country"],silver["Silver_Medal"])
plt.xticks(rotation = "vertical")
plt.xlabel("Country")
plt.ylabel("Value")
plt.show()
```

### Bronze Medals

```python
bronze = df[df["Bronze_Medal"]>0]

plt.figure(figsize=(15,8))
plt.bar(bronze["Country"],bronze["Bronze_Medal"])
plt.xticks(rotation = "vertical")
plt.xlabel("Country")
plt.ylabel("Value")
plt.show()
```

### Total Medals

```python
plt.figure(figsize=(20,10))
plt.bar(df["Country"],df["Total"])
plt.xticks(rotation = "vertical")
plt.xlabel("Country")
plt.ylabel("Value")
plt.show()
```
